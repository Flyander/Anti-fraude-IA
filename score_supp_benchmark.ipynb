```python
df_item = pd.DataFrame(columns=item_unique,index=df_train.index)
```

```python
for row in df_train.index:
    for i in range(1,25):
        if type(df_train[f"item{i}"][row]) == float: # Stop si valeur nulle
            break
        df_item[f"{df_train[f'item{i}'][row]}"][row] = df_train[f"Nbr_of_prod_purchas{i}"][row]
```

```python
df_make = pd.DataFrame(columns=make_unique,index=df_train.index)
```

```python
for row in df_train.index:
    for i in range(1,25):
        if type(df_train[f"make{i}"][row]) == float: # Stop si valeur nulle
            break
        df_make[f"{df_train[f'make{i}'][row]}"][row] = df_train[f"Nbr_of_prod_purchas{i}"][row]
```

```python
result = pd.concat([df_make, df_item], axis=1)
```

```python
for i in range(1,25):
  result[f'cash_price{i}'] = df_train[f'cash_price{i}']
  result[f'Nbr_of_prod_purchas{i}'] = df_train[f'Nbr_of_prod_purchas{i}']
```

```python
result['Nb_of_items'] = df_train['Nb_of_items']
```

```python
result = result.fillna(0)
```
